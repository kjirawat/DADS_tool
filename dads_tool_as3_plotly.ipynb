```python
import plotly.express as px
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/kjirawat/DADS_tool/81b3bb12c5fba5595745752a6340a1cf22712222/PS_PT_002_cleaned.csv')
df.head()
```

```python
mon_list = ["2023-Dec"]
df2 = df[df["Year-Month"].isin(mon_list)]
print(df2)
```

```python
fig = px.sunburst(df2, path=['Year-Month','Payment Type', 'Sub Type'], values='Volume (x1000)',
                 width=1050, height=1050, color="Volume (x1000)")
fig.show()
```
